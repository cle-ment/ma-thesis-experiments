# Hyper-paramter Analysis

```python
def plot_training(scores, 
                  labels=["Train", "Test", "Train inferred", "Test inferred"], 
                  size=(3.4,3.4), filename=None, title=None, xlim=140):
    
    with sns.axes_style("whitegrid"):
        with sns.color_palette("hls", len(scores)):
            plt.figure(figsize=size)
            
            for i, score in enumerate(scores):
                plt.plot(np.arange(len(score)), score, label=labels[i]) 

            plt.xlabel('Training Epoch')
            plt.ylabel("MCC")
            
            axes = plt.gca()
            axes.set_ylim(0,1)
            axes.set_xlim(0,xlim)

            legend = plt.legend(loc='upper right', frameon=True)
            legend.get_frame().set_alpha(0.7)
            frame = legend.get_frame()
            frame.set_facecolor('#FFFFFF')
            
            plt.tight_layout() 
            
            if title:
                plt.title(title)

            if filename:
                plt.savefig("./plots/doc2vec_" + filename +".pdf", format='pdf')

            plt.show()  
```
